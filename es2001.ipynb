```python
import pandas as pd

vendite=pd.read_csv("vendite.csv")

#print(vendite.head())
#print(vendite.info())
#print(vendite.describe())

#vendite_grandi=vendite[vendite["PrezzoUnitario"]>1000]

vendite["TotaleVendita"]=vendite["PrezzoUnitario"]*vendite["Quantita"]
vendite["Commissione"]=vendite["TotaleVendita"]*0.10

vendite["id_agente"]=[
  6, 3, 9, 1, 10, 4, 7, 2, 8, 5, 10, 1, 6, 3, 9, 2, 7, 4, 8, 5,
  1, 10, 4, 6, 3, 9, 2, 8, 5, 7, 4, 1, 10, 6, 3, 2, 9, 8, 5, 7,
  9, 2, 1, 10, 6, 3, 7, 4, 8, 5, 2, 9, 6, 3, 10, 1, 8, 5, 7, 4,
  3, 6, 9, 2, 1, 10, 5, 8, 4, 7, 6, 3, 10, 1, 2, 9, 8, 5, 7, 4,
  10, 1, 6, 3, 9, 2, 7, 4, 8, 5, 1, 10, 4, 6, 3, 9, 2, 8, 5, 7
]

per_venditore=vendite.groupby('id_agente').agg({
    'TotaleVendita': 'sum',
    'Commissione': 'mean',
    'id_agente': 'count'
})
per_venditore.columns=['TotaleVendite', 'TotaleCommissioni', 'Numero']

per_venditore=per_venditore.sort_values(by='TotaleVendite', ascending=False)
per_venditore.to_csv("report_venditori.csv")

per_venditore

```
